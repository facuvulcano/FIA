Unicode

Concepto de Code-Point

Cada caracter en el estandar Unicode se representa mediante un numero conocido como "code-point". Este numero es unico para cada caracter y permite representar caracteres de casi todos los sistemas de escritura del mundo. Unicode tiene aproximadamente 150.000 code-points.

Codificacion de Unicode

Unicode define varios metodos de codificacion para convertir estos code-points en una secuencia de bytes que pueden ser almacenados o transmitidos.

1- UTF-8 (8-bit Unicode Transformation Format)

- Longitud Variable: Puede usar de 1 a 4 bytes por caracter.

- Ventaja: Compatible con ASCII, lo que lo hace muy eficiente para textos en ingles y muy popular en la web.

- Ejemplo: El caracter 'A' (U + 0041) se codifica como 0x41 en UTF-8

2- UTF-16 (16-bit unicode Transformation Format)

- Longitud Variable: Usa 2 o 4 bytes por caracter.

- Ventaja: Mas eficiente para textos que icluyen muchos caracteres fuera del rango basisco multilingue (BMP).

- Ejemplo: El caracter 'A' (U + 0041) se codifica como 0x0041 en UTF-16.

3- UTF-32 (32-bit Unicode Transformation Format)

- Longitud Fija: Usa 4 bytes por caracter.

- Ventaja: Cada caracter se representa por el mismo numero de bytes, lo que simplifica el procesamiento.

- Ejemplo: El caracter 'A' (U + 0041) se codifica como 0x00000041 en UTF-32

Estas codificaciones permiten que los caracteres Unicode se almacenen y transmitan de manera eficiente, manteniendo la compatibilidad con diferentes sistemas y plataformas.


Representacion de Textos en NLP: Caracteres vs. Palabras

Cuando representamos textos para tareas de NLP, tenemos dos opciones principales: utilizar caracterse o utilizar palabras. Cada enfoque tiene sus propias ventajas y desafios.

Representacion por Caracteres

- Conjunto de Caracteres: En este enfoque, el texto se divide en sus caracteres individuales. Cada caracter se representa numericamente utilizando su code-point Unicode.

- Ventajas:

        - No requiere segmentacion previa del texto.
        - Puede capturar detalles finos de la ortografia y la gramatica.
        - Util para lenguajes con escritura compleja y morfologia rica.

- Desafios:

        - Requiere una gran cantidad de datos de entrenamiento debido a la longitud de las secuencias.
        - El modelo necesita aprender dependendencias a largo plazo, lo cual es complejo
        - Menor eficiencia en terminos de procesamiento y almacenamiento.


Representacion por Palabras

- Conjunto de palabras (Tokens): En este enfoque, el texto se divide en palabras (tokens). Cada palabra se representa mediante un indice en un vocabulario predefinido.

- Ventajas:

        - Reduccion en la longitud de las secuencias, lo que facilita el entrenameinto.
        - Captura mas significado semantico a nivel de palabra
        - Comunmente utilizado en muchas aplicaciones de NLP.

- Desafios:

        - Necesita un proceso de tokenizacion que puede ser complejo para algunos lenguajes.
        - Manejo de palabras fuera del vocabulario (OOV - Out Of Vocabulary)


Tokenizacion en NLP

La tokenizacion es el proceso de convertir un texto en una secuencia de unidades significativas llamadas tokens. Es una parte fundamental de todos los problemas de procesamiento del lenguaje natural (NLP), y su correcta implementacion es crucial para el rendimiento de cualqueir modelo de lenguaje, incluyendo GPT y otros modelos de aprendizaje profundo.

Importancia de la tokenizacion

1- Base de Todo el NLP: La tokenizacion es la primera etapa en la mayoria de los sistemas de NLP. es el paso necesario que convierte un texto bruto en una forma que puede ser procesada por modelos de aprendizaje automatico.

2- Impacto en el rendimiento: Una tokenizacion inapropiada puede degradar significativamente el rendimiento de un modelo, ya que afecta directamente como se interpretan y representan las palabras y frases.

3- Problema intrinseca: Aunque esencial, la tokenizacion es imperfecta y la raiz de muchos problemas. Diferentes lenguajes y aplicaciones presentan desafios unicos, y las decisiones de disenio en la tokenizacion pueden tener efectos de largo alcance.

Metodos basicos y desafios

- Separacion por espacios:
        
        - Ventajas: Simple y rapida
        - DesventajasL No maneja correctamente palabras compuestas, puntuacion o lenguajes donde los espacios no delimitan palabras (como el chino o el japones).

Problemas de Tokenizacion

1- Perdida de espacios importantes: En ciertos contextos, como en el codigo fuente de programacion (e.g., Python), los espacios son cruciales y su eliminacion o alteracion puede cambiar el significado o incluso invalidar el codigo.

2- Palabras compuestas:
        
        - Ejemplo en ingles: "New York" podria ser tratado como un solo token o dos tokens separados ("New" y "York").
        - Implicaciones: La eleccion puede afectar la precision de la interpretacion del modelo.

3- Manejo de puntuacion:

        - Problema: Si no se maneja correctamente, puede resultar en tokens inapropiados.

4- Lenguaje sin espacios:

        - Ejemplo: Chino y Japones no usan espacios para delimitar palabras, lo que hace la tokenizacion mas compleja.
        - Soluciones: Algoritmos de segmentacion especificos para estos lenguajes.

Algoritmo mas populas: BPE

Byte Pair Encoding (BPE) es una tecnica popular para la tokenizacion utilizada en muchos modelos de NLP modernos. Es especialmente util para reducir el tamanio del vocabulario y manejar palabras raras o fuera del vocabulario (OOV).

Funcionamiento de BPE

1- Conversion a UTF-8:

        - Paso inicial: Convertir el texto en una secuencia de bytes utilizando UTF-8. Esto garantiza que cada caracter se represente como una secuencia de bytes.

        - Ejemplo:

        Texto: "Hoy es lunes y hace fro es importante abrigarse."
        UTF-8 Bytes (Ejemplo simplificado): [72, 111, 121, 32, 101, 115, 32, 108, 117, 110, 101, 115, 32, 121, 32, 104, 97, 99, 101, 32, 102, 114, 105, 111, 32, 101, 115, 32, 105, 109, 112, 111, 114, 116, 97, 110, 116, 101, 32, 97, 98, 114, 105, 103, 97, 114, 115, 101]

2- Encuentra el par de bytes mas frecuente:

        - Paso: Identificar el par de bytes adyacentes mas frecuente en la secuencia.

        - Ejemplo: Supongamos que 'es' es el par mas frecuente.

3- Crear un nuevo token:

        - Paso: Reemplazar todas las apariciones del par mas frecuente con un nuveo token.

        - Ejemplo: Reemplazar todas las apariciones de 'es' como un nuevo token, digamos 256

4- Repetir el proceso:

        - Paso: Continuar el proceso iterativamente hasta alcanzar un numero deseado de tokens.

        - Ejemplo: Despues de 'es', podriamos encontrar que 32 256 (espacio seguido de es) es el mas frecuente, y crear un nuevo token, digamos 257, para 'espacio es'



Una vez que hemos creado estos tokens, enfrentamos un problema conocido: la representacion numerica de estos tokens en una red neuronal no tiene un significado intrinseco. La representacion directa de los tokens por sus codigos numericos puede ser problematica al utilizarlos en una red neuronal.

Solucion: Embeddings.

Los embeddings son representaciones vectoriales de palabras que permiten capturar sus relaciones semanticas en un espacio de alta dimension. Estos embeddings se entrenan de manera que palabras similares en significado esten cerca en este espacio vectorial.

Definicion de Embeddings

        - Vector de Dimensiones: Cada palabra (token) se representa mediante un vector de n dimensiones. La cantidad de dimensiones n suele ser un hiperparametro, comunmente entre 50 y 300 dimenesiones.

Calculo de la semejanza entre palabras

Para medir la semejanza entre dos palabras, se utiliza comunmente la distancia coseno. Esta medida de semejanza es especialmente adecuada para embeddings, ya que captura la orientacion de los vectores en el espacio de alta dimension, sin importar su magnitud.

Distancia coseno

La distancia coseno entre dos vectores X e Y se define como:

        - cos tita = (x . y ) / (|x| . |y|)

Donde:

        - x . y es el producto interno de los vectores x e y.
        - |x| es la norma (magnitud) del vector x
        - |y| es la norma (magnitud) del vector y

Normalizacion

Si los vectores x e y estan normalizados (es decir, tienen una norma de 1), la formula se simplifica a:

cos tita = x . y

Esto significa que la semejanza coseno se reduce al producto interno de los dos vectores.





Interpretacion de la semejanza coseno

- Valor entre -1 y 1: la semejanza coseno varia entre -1 y 1.

        - 1: los vectores son identicos en direccion
        - 0: Los vectores son ortogonales (sin relacion)
        - -1: Los vectores son diametralmente opuestos.


Finalmente usamos una Softmax:

La capa softmax se utiliza comunmente en tareas de clasificacion. Proporciona una distribucion de probabilidad sobre las posibles clases de salida, lo cual es esencial para muchas aplicaciones de NLP, como la clasificacion de texto, el etiquetado de secuencias, y la traduccion automatica.


Flujo completo de un modelo de NLP

1- Tokenizacion:

        - El texto se divide en tokens, a menudo utilizando tecnicas como Byte Pari Encoding (BPE).

2- Embeddings:

        - Los tokens se convierten en vectores densos de alta dimension mediante una capa de embedding. Esta capa puede estar preentrenada (como Word2Vec, GloVe, o embeddings de BERT) o entrenarse conjuntamente con el modelo.

3- Procesamiento Secuencial:

        - Los embeddings se pasan a traves de una red neuronal, uqe puede una RNN, LSTM, GRU, Transformer, etc. Este paso permite al modelo capturar dependencias contextuales y secuenciales entre los tokens.

4- Capa Densa y Softmax:

        - Finalmente, el resultado del procesamiento secuencial pasa por una 0 varias capas densas, y luego se utiliza una capa Softmax para producir una distribucion de probabilidad sobre las clases de salida.


Ejemplo: Clasificacion de Texto

Supongamos que estamos entrenando un modleo para clasificar el sentimiento de un texto (positivo, negativo, neutral). Aqui esta el flujo completo:

1- Tokenizacion:

        - Texto: "Me encanta este producto"
        - Tokens: ["Me", "Encanta", "este", "producto", "."]

2- Embbeddings:

        - Tokens convertidos a vectores de embedding
        - Ejemplo: ["Me" -> [0.1, 0.3, ...], "encanta -> [0.4, 0.5,...], ...]

3- Procesamiento Secuencial:

        - Los embeddings se pasan a traves de una red neuronal (ej., LSTM o Transformer) para capturar el contexto.

4- Capa densa:

        - El resultado de la red secuencial se pasa a traves de una o mas capas densas.
        - Ejemplo: La salida de la LSTM podria ser un vector de 256 dimensiones, que se pasa a una capa densa de 3 unidades (correspondientes a las clases de salida: positivo, negativo, neutral).

5- Softmax:

        - La salida de la capa densa se pasa a una capa Softmax, que convierte los logits en probabildiades.
        - Ejemplo: [1.2, -0.8, 0.3] se convierte en [0.65, 0.10, 0.25] (probabilidades de las clases)

Word2Vec es una tecnica de procesamiento del lenguaje natural desarrollada por un equipo de Google, que transforma las palabras en vectores numericos de una manera que captura los significados contextuales y las relaciones semanticas entre las palabras.

Word2Vec tiene dos arquitecturas princiaples:

1- CBOW (Continuos Bag of Words): Predice una palabra basada en su contexto

2- Skip-gram: Predice el contexto basado en una palabra dada.

Proceso de Entrenamiento

1- Inicializacion:

- Las palabras del vocabulario se inicializan con vectores aleatorios

2- Contexto y Ventana Deslizante:

- Se define una ventana de contexto de tamanio n. Por ejemplo, si n = 2, para la palabra rapido en la oracion. "El zorro rapido salta sobre el perro perezoso", el contexto seria ["El", "zorro", "salta", "sobre"].

3- Actualizacion de vectores:

- En el caso de CBOW, el modelo toma el contexto ("El", "zoro", "salta", "sobre") y predice la palabra objetivo ("rapido").

- En el caso de Skip-gram, toma la palabra objetivo ("rapido") y trata de predecir las palabras del contexto ("El", "zorro", "salta", "sobre").

- Los vectores se ajustan mediante un proceso de retropropagacion y optimizacion (normalmente usando el gradiente descendente).

Ejemplo simplificado

Supongamos que tenemos la oracion "El zorro rapido salta sobre el perro perezoso".

Skip-gram:

- Objetivo: dada una palabra central, predecir sus palabras de contexto
    - Para la palabra "rapido":
        - contexto: "El", "zorro", "salta", "sobre"
        - Entrenamientos: (rapido, El), (rapido, zorro), (rapido, salta), (rapido, sobre)

CBOW:

- Objetivo: dado un contexto, predecir la palabra central.
    - Para el contexto "El zoro salta sobre":
        - Objetivo: "Rapido"
        - Entrenamiento: (El, zorro, salta, sobre) -> "rapido".

Vectores Resultantes

Despues del entrenamiento, cada palabra se representa como un vector en un espacio de alta dimension (por ejemplo, 100, 200, 300 dimensiones). Las palabras que tienen significados y contextos similares estaran cerca en este epacio vectorial. Por ejemplo, los vectores de "rey" y "reina" estaran cerca entre si y tendran una relacion matematica similar a la de "hombre" y "mujer".

Usos de Word2Vec

1- Deteccion de sinonimos: Las palabras similares tendran vectores similares

2- Analogias de palabras: Por ejemplo, el modelo puede resolver "rey - hombre + mujer = reina".

3- Mejora de modelos de lenguaje: Los vectores pueden ser utilizados como caractersticas en modelos mas complejos para tareas como la traduccion automatica, la generacion de texto, etc.

Language models

Los modelos de lenguaje, especificamente los modelos basados en la arquitectura Transformer, se dividen principalmente en dos tipos: Encoder-Decider y Decoder-only.

Modelos Encoder-Decoder

Estan diseniados principalmente para tareas de secuencia a secuencia (seq2seq), como la traduccion automatica, el resumen de texto, entre otros. La arquitectura consta de dos partes:

1- Encoder: procesa la secuencia de entrada y la convierte en una representacion de contexto

2- Decoder: Genera la secuencia de salida utilizando la representacion del contexto proporcionada por el encoder.

Entrenamiento

Durante el entrenamiento, el modelo reibe una secuencia de entrada y su correspondiente secuencia de salida. Por ejemplo, en una tarea de traduccion:

- Entraada (input): "Hoy es viernes"
- salida esperada (ouput): "<start>today is friday<end>"

Las secuencias de entrada y salida se ajustan para tener la misma longitud utilizando tokens espeicales como <pad> para el relleno si es necesario

- Entrada: "Hoy es viernes <pad> <pad>"
- Salida: "<start> today is friday <end>"

El modelo se entrena para minimizar la diferencia entre la secuencia generada y la secuencia esperada, token por token.

Prediccion

Durante la prediccion, el proceso es incremental. Se utliza la secuencia de entrada y se genera la secuencia de salida token por token:

- Entrada inicial: "Hoy es viernes <pad> <pad>"
- Salida inicial: "<start>"

El modelo predice el primer token de salida

- Entrada: "Hoy es viernes <pad> <pad>"
- Salida: "<start> today"

Luego, el modelo usa el token generado ("today") para predecir el siguiente token:

- Entrada: "Hoy es viernes <pad> <pad>"
- Salida: "<start> today is"

Este proceso continua hasta que se genera el token final <end>.

Representacion y calculo

- Entradas: La entrada al modelo Encoder-Decoder es un tensor de dimensiones (context_size x model_dimension)

- Salidas: La salida es una matriz de probabildiades generada por una capa softmax, que tiene dimensiones (context_size x vocab_size), donde vocab_size es el numero de tokens unicos en el vocabulario.

Cada token de la salida se genera uno por uno utilizando la distribucion de probabilidad obtenida del Softmax para elegir el token mas probable en cada paso.

Ventjaas

- Capacidad de contexto: Los modelos Encoder-Decoder pueden manjear relaciones complejas en el texto de entrada y salida, considerando contexto complejo de la secuencia.

- Flexibilidad: Son aplicables a varias tareas de secuencia a secuencia, no solo traduccion, sino tambien resumen, generacion de texto, entre otros.

Componentes principales del Transformer

La arquitectura Transformer se compone de dos partes principales: El encoder y el decoder. Ambas partes estan formadas por una pila de capas de redes neuronales identicas.

Encoder

El encoder esta compuesta por una pila de N capas identicas. Cada capa tien dos subcapas principales:

1- Multi-Head Self-Attention Mechanism:

- Petmite que cada posicion en la secuencia de entrada se enfoque en otras posiciones para obtener una representacion mas rica.

2- Feed-Forward Neural Network:

- Aplicada de forma independiente a cada posicion de la secuencia.

Ademas, se utilizan tecnicas de normalizacion y residual connections para mejorar el flujo de gradientes durante el entrenamiento.

Decoder

El Decoder tambien esta compuesto por una pila de N capas identicas, pero cada capa tiene una subcapa adicioal para manejar la atencion sobre la salida del Encoder.

1- Masked Multi-Head Self-Attention Mechanism:

- Similar al mecanismo de atencion del encoder, pero enmascara las posiciones futuras en la secuencia de salida para asegurar que la prediccion en un tiempo t solo dependa de las posiciones anteriores.

2- Multi-Head Attention over the Encoder's Output:

- Permite que el Decoder se enfoque en las partes relevantes de la salida del encoder.

3- Feed-Forward Neural Network

- Simiar a la subcapa en el Encoder.
