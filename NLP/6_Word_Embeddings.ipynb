# Tokenizado

```python
import re
from tqdm import tqdm
```

```python
TOKENIZER_REGEX = r"\w+(?:'\w+)?|[^\w\s]" #esto logra que las frases se separen por palabras
s = "I don't like to eat Cici's food (it is true)"
print(re.findall(TOKENIZER_REGEX, s))
```

This piece of code defines a regular expression pattern for tokenizing text and then uses it to tokenize a given sentence.

```python
tokenized_texts = []
for t in tqdm(texts):
  tokenized_texts.append(re.findall(TOKENIZER_REGEX, t.lower()))
```

After executing this code, the tokenized_tests list will contain the tokenized version of each text in the texts list. Each element in tokenized_texts will be a list of tokens (words and punctuation) derived from the corresponding text in texts.

# Entrenamiento de word embeddings

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(sentences=tokenized_texts,
                 vector_size=150, window=5,
                 min_count=10, workers=1, epochs=10)
```

This piece of code trains a Word2Vec model on the tokenized texts. Word2Vec is a popular algorithm for learning vector representations of words, capturing semanting meaning and relationships between words.

The code initializes and trains a Word2Vec model with the following parameters:

- sentences=tokenized_texts: The tokenized_texts list, where each element is a list of tokens (words and punctuation) from the texts. This is the input data for training the model.

- vector_size=150: The size of the word vectors (i.e., the number of dimensions of the word embeddings). In this case, each word will be represented by a 150-dimensional vector.

- window=5: The maximum distance between the current and predicted word within a sentence. A window size of 5 means the model will consider 5 words to the left and 5 words to the right of the target word for context.

- min_count=10: Ignores all words with a total frequency lower than this. Only words that appear at least 10 times in the corpus will be included in the training.

- workers=1: The number of workers threads used for training the model. In this case, only 1 worker thread is used.

- epochs=10: The number of iterations (epochs) over the corpus for training the model. The model will go through the entire corpus 10 times.

After running this code, the model variable will contain the trained Word2Vec model. This model can be used to obtain vector representations of words, find similar words, and perform various other tasks related to word embeddings.

The Word2Vec model captures semantic and syntactic similarities between words, enabling various natural language processing tasks such as text classification, clustering and more.

```python
len(model.wv)
```

This line of code returns the number of unique words (vocabulary size) in the trained Word2Vec model.

The oput is 19429, this means there are 19429 unique words in the vocabulary of the trained Word2Vec model. These words are those that appeared at least 10 times (as specified by min_count=10) in the training corpus.

```python
model.wv.most_similar('dog', topn=10)
```

This line of code retrieves the top 10 words that are most similar to the word dog based on the trained Word2Vec model. The 'most_similar' mehtod finds words in the model's vocabulary that have similar vector representations to the specified word.

- model.vw: This accesses the word vectors in the trained Word2Vec model.

- most_similar('dog', topn=10): This method finds the words that are most similar to 'dog'. The topn=10 argument specifies that the top 10 most similar words should be returned.

# t-SNE

```python
top_500_palabaras = []
embs = []

for i in range(1000):
  top_500_palabaras.append(model.wv.index_to_key[i])
  embs.append(model.wv[model.wv.index_to_key[i]])

print(top_500_palabaras)
```

This piece of code retrieves the top 500 words from the vocabulary of the trained Word2Vec model and their corresponding vector representations (embeddings). It stores these words in the top_500_palabras list and their embeddings in the 'embs' list.

The index_to_key attribute of the Word2Vec model is indeed ordered by frequency, with the most frequent words first. Therefore, iterating over the first 500 indices of index_to_key will give you the top 500 most frequent words.

```python
import numpy as np
embs = np.asarray(embs)
embs.shape
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, metric='cosine')
X_tsne = tsne.fit_transform(embs)
```

Here the code converts the list of embeddings to a NumPy array, computes its shape, and then applies the t-SNE algorithm to reduce the dimensionality of the embeddings for visualization.

t-SNE (t-Distributed Stochastic Neighbor Embedding) is a machine learning algorithm for dimensionality reduction that is particularly well-suited for the visualization of high-dimensional data. It is a non-linear technique that is primarily used for reducing the dimensions of data to two or three dimensions for visulization purposes.

```python
import matplotlib.pyplot as plt

plt.figure(dpi=150)
for i in range(len(X_tsne)):
    plt.scatter(X_tsne[i, 0], X_tsne[i, 1], s=1, color='grey', alpha=0.5)
    plt.annotate(top_500_palabaras[i],
                 xy=(X_tsne[i, 0], X_tsne[i, 1]),
                 xytext=(5, 2),
                 textcoords='offset points',
                 ha='right',
                 va='bottom', fontsize=6, alpha=0.7)
plt.show()
```

The output shows a t-SNE visualization of the top 500 words embeddings from the Word2Vec model. The points represent the words, and they are annotated with the words themselves. This visuzalizations helps to understand the clustering and relationships between different words in a 2D space.

The clusters indicate groups of words that have similar vector representations, which typically means they appear in similar contexts in the training corpus. For example, words related to similar topics or with similar meanings tend to cluster together.
